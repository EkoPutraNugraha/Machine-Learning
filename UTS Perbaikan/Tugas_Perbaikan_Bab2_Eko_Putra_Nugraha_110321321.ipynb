```python
# Import pustaka yang diperlukan
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification, make_regression
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, mean_squared_error
```

```python
# 1. Dataset untuk Klasifikasi
print("=== Dataset Klasifikasi ===")
X_class, y_class = make_classification(
    n_samples=200,
    n_features=2,
    n_informative=2,  # Semua fitur adalah informatif
    n_redundant=0,    # Tidak ada fitur redundan
    n_classes=2,
    random_state=42
)

# Menampilkan contoh data
print("Contoh data klasifikasi (5 baris pertama):")
print(X_class[:5])
print("Target (5 baris pertama):", y_class[:5])

# Memisahkan data menjadi data latih dan uji
X_train_class, X_test_class, y_train_class, y_test_class = train_test_split(
    X_class, y_class, test_size=0.3, random_state=0
)
```

Kode dan output di atas menggunakan fungsi make_classification untuk membuat dataset simulasi yang digunakan dalam tugas klasifikasi. Dataset ini terdiri dari 200 sampel dengan dua fitur (n_features=2), yang semuanya bersifat informatif (n_informative=2) tanpa fitur redundan (n_redundant=0). Dataset juga dibagi menjadi dua kelas target (n_classes=2), di mana setiap sampel diberi label target 0 atau 1. Pengaturan parameter random_state=42 memastikan bahwa hasil pembuatan dataset akan konsisten setiap kali kode dijalankan. Lima sampel pertama dari dataset menunjukkan nilai-nilai fitur yang dihasilkan secara acak tetapi tetap sesuai dengan pola yang memisahkan dua kelas.

Setelah dataset dibuat, data dibagi menjadi data latih dan data uji menggunakan fungsi train_test_split dengan proporsi 70% untuk data latih dan 30% untuk data uji (test_size=0.3). Data latih akan digunakan untuk melatih model klasifikasi, sedangkan data uji akan digunakan untuk mengevaluasi kinerja model tersebut. Lima baris pertama dari data fitur menunjukkan nilai numerik acak yang membentuk dua dimensi fitur, sementara target (label) untuk lima sampel pertama adalah [1, 1, 0, 0, 1], yang menunjukkan kelas dari masing-masing sampel. Pemisahan ini merupakan langkah penting untuk memastikan evaluasi model yang adil dan mengukur kemampuan

```python
# 2. Klasifikasi dengan KNN
print("\n=== Klasifikasi dengan KNN ===")
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train_class, y_train_class)

# Prediksi dan evaluasi
y_pred_class = knn.predict(X_test_class)
accuracy_class = accuracy_score(y_test_class, y_pred_class)
print("Akurasi pada data uji: {:.2f}".format(accuracy_class))

# Visualisasi hasil klasifikasi
plt.scatter(X_test_class[:, 0], X_test_class[:, 1], c=y_pred_class, cmap='viridis', marker='o')
plt.title("Hasil Klasifikasi dengan KNN")
plt.xlabel("Fitur 1")
plt.ylabel("Fitur 2")
plt.show()
```

Model K-Nearest Neighbors (KNN) diterapkan pada dataset klasifikasi yang telah dibuat untuk memprediksi kelas dari data uji. Dengan menggunakan parameter n_neighbors=3, model mempertimbangkan tiga tetangga terdekat untuk menentukan label kelas suatu sampel. Model dilatih menggunakan data latih (X_train_class, y_train_class) dan menghasilkan prediksi untuk data uji (X_test_class). Evaluasi menunjukkan bahwa model mencapai akurasi 90% pada data uji, yang berarti model berhasil memprediksi dengan benar untuk 90% sampel dalam data uji. Tingkat akurasi yang cukup tinggi ini menunjukkan bahwa model KNN mampu menangkap pola yang relevan dalam dataset latih untuk membedakan dua kelas.

Visualisasi hasil klasifikasi menunjukkan distribusi sampel data uji berdasarkan dua fitur (Fitur 1 dan Fitur 2) dengan warna yang menunjukkan kelas hasil prediksi model. Titik-titik berwarna kuning dan ungu merepresentasikan dua kelas target yang berbeda. Sebagian besar titik-titik tampak terklasifikasi dengan baik sesuai pola kelasnya, meskipun mungkin ada beberapa sampel yang terklasifikasi salah (tidak terlihat langsung dalam grafik ini). Scatter plot ini memberikan gambaran visual yang membantu memahami bagaimana model KNN membagi ruang fitur berdasarkan prediksi kelasnya, yang selaras dengan akurasi tinggi yang dicapai model.

```python
# 3. Dataset untuk Regresi
print("\n=== Dataset Regresi ===")
X_reg, y_reg = make_regression(n_samples=200, n_features=1, noise=20, random_state=42)

# Menampilkan contoh data
print("Contoh data regresi (5 baris pertama):")
print(X_reg[:5])
print("Target (5 baris pertama):", y_reg[:5])

# Memisahkan data menjadi data latih dan uji
X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(
    X_reg, y_reg, test_size=0.3, random_state=0
)
```

Kode di atas menggunakan fungsi make_regression untuk menghasilkan dataset simulasi yang digunakan dalam tugas regresi. Dataset ini terdiri dari 200 sampel dengan satu fitur (n_features=1) dan target nilai kontinu (y_reg). Parameter noise=20 ditambahkan untuk menciptakan variasi acak pada data, sehingga mencerminkan situasi nyata di mana data biasanya memiliki beberapa derajat kebisingan. Lima sampel pertama dari data input (X_reg) menunjukkan nilai numerik dari fitur, sementara target (y_reg) merepresentasikan nilai keluaran kontinu yang sesuai. Target ini dihitung berdasarkan hubungan linier antara fitur dan target, dengan tambahan noise.

Setelah dataset dibuat, data dibagi menjadi data latih dan data uji menggunakan fungsi train_test_split, dengan 70% data dialokasikan untuk pelatihan dan 30% untuk pengujian. Data latih akan digunakan untuk melatih model regresi, sedangkan data uji akan digunakan untuk mengevaluasi kinerja model tersebut. Pembagian ini penting untuk memastikan model dapat diuji pada data yang tidak terlihat selama pelatihan, sehingga memberikan evaluasi yang adil tentang kemampuan generalisasi model. Hasil ini menunjukkan langkah awal yang umum dilakukan dalam tugas regresi, yaitu menghasilkan data simulasi, memeriksa struktur data, dan mempersiapkan data untuk proses pelatihan dan evaluasi model.

```python
# 4. Regresi dengan Linear Regression
print("\n=== Regresi dengan Linear Regression ===")
lr = LinearRegression()
lr.fit(X_train_reg, y_train_reg)

# Prediksi dan evaluasi
y_pred_reg = lr.predict(X_test_reg)
mse_reg = mean_squared_error(y_test_reg, y_pred_reg)
print("Mean Squared Error pada data uji: {:.2f}".format(mse_reg))

# Visualisasi hasil regresi
plt.scatter(X_test_reg, y_test_reg, label="Data Uji", color="blue")
plt.plot(X_test_reg, y_pred_reg, label="Prediksi", color="red")
plt.legend()
plt.title("Hasil Regresi dengan Linear Regression")
plt.xlabel("Fitur")
plt.ylabel("Target")
plt.show()
```

Hasil output menunjukkan implementasi regresi linier sederhana pada dataset yang dihasilkan sebelumnya. Model LinearRegression dari pustaka scikit-learn dilatih menggunakan data latih (X_train_reg dan y_train_reg). Setelah pelatihan, model digunakan untuk memprediksi target nilai kontinu pada data uji (X_test_reg). Evaluasi model dilakukan dengan menghitung Mean Squared Error (MSE), yang menghasilkan nilai 416.31. Nilai MSE ini merepresentasikan rata-rata kuadrat dari kesalahan prediksi, dengan nilai yang lebih kecil menunjukkan prediksi model yang lebih akurat.

Visualisasi hasil regresi menunjukkan hubungan antara fitur dan target. Titik-titik biru pada scatter plot merepresentasikan data uji sebenarnya, sementara garis merah menunjukkan prediksi model. Grafik ini menunjukkan bahwa garis regresi model cukup baik dalam mendekati data uji, meskipun terdapat beberapa deviasi antara prediksi dan nilai target aktual, yang mencerminkan adanya noise yang telah ditambahkan sebelumnya dalam dataset. Hasil ini menunjukkan bahwa regresi linier mampu menangkap pola utama dari data, meskipun performa model dipengaruhi oleh keberadaan noise dalam data.
